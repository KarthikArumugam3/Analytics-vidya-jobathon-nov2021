```python
# made new column named target
train["target"] = train["LastWorkingDate"].fillna(0)
train["target"] = train['target'].apply(lambda x : 1 if(x!= 0) else 0)
```

```python
# made new column named NumOfMonths_empWorked
a = train["Emp_ID"].value_counts()

def for_new_col(x):
    a = train["Emp_ID"].value_counts()
    return a[x]
```

```python
train['NumOfMonths_empWorked'] = train['Emp_ID'].apply(lambda x: for_new_col(x))
```

```python
# made new column named Mean_ratings
train['Mean_ratings'] = train['Emp_ID'].apply(lambda x: train[train['Emp_ID'] == x]["Quarterly Rating"].values.mean())
```

```python
# made new col named zero_val_gen
# this column implies number of months when the total sales is less than equal to zero

train['zero_val_gen'] = train['Emp_ID'].apply(lambda x: train[(train['Emp_ID'] == x) & (train["Total Business Value"]<=0)].shape[0])
```

```python
# made new column named Mean_business_val
train['Mean_business_val'] = train['Emp_ID'].apply(lambda x: train[train['Emp_ID'] == x]["Total Business Value"].values.mean())
```

```python
# made new column for change in designation
train['chg_in_desg'] = train['Designation']-train['Joining Designation']
```

```python
# total business value by employee to company
train['ttl_busn_val'] = train['Emp_ID'].apply(lambda x: train[train['Emp_ID'] == x]["Total Business Value"].values.sum())
```

```python
# made new column for Age Group category, so that we can easily make dummies and drop Age column
bins_for_AgeGroup = [-1,0,20,25,29,32,35,40,45,51, 110]
labels_for_AgeGroup = ['<0','0_19','20_24','25_28','29_31','32_34','35_39','40_44','45_50', '51_110']
train['AgeGroup'] = pd.cut(train['Age'], bins=bins_for_AgeGroup, labels=labels_for_AgeGroup, right=False)
```

```python
# made new column for Salary group category, so that wee can easily make dummies and drop Salary column
bins_for_SalaryGroup= [0,21000,50000,80000,100000,120000,150000,180000,500000]
labels_for_SalaryGroup = ['<21000','21k_50k','50k_80k','80k_1L','1L_1.2L','1.2L_1.5L','1.5L_1.8L','1.8L_5L']
train['SalaryGroup'] = pd.cut(train['Salary'], bins=bins_for_SalaryGroup, labels=labels_for_SalaryGroup, right=False)
```

```python
bins_for_ttl_busn_val_grp= [-7000000,-5000000,-1000000,0,100000,1000000,2000000,5000000,7000000,9000000,50000000]
labels_for_ttl_busn_val_grp = ['-70L_-50L','-50L_-10L','-10L_0','0_1L','1L_10L','10L_20L','20L_50L','50L_70L','70L_90L','90L_5Cr']
train['ttl_busn_val_grp'] = pd.cut(train['ttl_busn_val'], bins=bins_for_ttl_busn_val_grp, labels=labels_for_ttl_busn_val_grp, right=False)
```

```python
# making new col of month and year for date of joining
train['Dateofjoining'] = pd.to_datetime(train['Dateofjoining'])
```

```python
train['Year_d_o_Joining'] = train['Dateofjoining'].dt.year
train['month_d_o_Joining'] = train['Dateofjoining'].dt.month
```

```python
bins_for_month_d_o_Joining= [1,4,7,10,13]
labels_for_month_d_o_Joining = ['jan_mar','apr_jun','jul_sep','oct_dec']
train['month_d_o_Joining'] = pd.cut(train['month_d_o_Joining'], bins=bins_for_month_d_o_Joining, labels=labels_for_month_d_o_Joining, right=False)
```

```python
train['Year_d_o_Joining'] = train['Year_d_o_Joining'].astype('str')
```
